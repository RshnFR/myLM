```python
from datasets import load_dataset

ds = load_dataset("stanfordnlp/sst2")
```

```python
print(ds)
```

```python
def tf_idf(dataset):
    """ 
    term frequency-inverse document frequency
    """
    bag_of_words = set([word for sentence in dataset['test']['sentence'] for word in sentence.split()])
    tf = {word: 0 for word in bag_of_words}
    idf = {word: 0 for word in bag_of_words}
    for sentence in dataset['test']['sentence']:
        for word in sentence.split():
            tf[word] += 1
        unique_words = set(sentence.split())
        for word in unique_words:
            idf[word] += 1
    for word in bag_of_words:
        tf[word] /= len(dataset['test']['sentence'])
        idf[word] = len(dataset['test']['sentence']) / idf[word]
    tf_idf = {word: tf[word] * idf[word] for word in bag_of_words}
    return tf_idf
tf_idf_values = tf_idf(ds)
```
